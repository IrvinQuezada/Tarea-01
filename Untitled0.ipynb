```
#Ejercicio 5 
p=150
m=170
print("Este programa hace el calculo de cuanto pesa su pedido en KG")
print("*Recuerda que tu pedido no debe sobrepasar los 10 KG*")

x = int(input("¿Cuantos payasos comprará? "))
y = int(input("¿Cuantos juegos de mesa comprará? "))

z= (p/1000*x)+(m/1000*y)
if z<10:
    print("El peso de su pedido es:",z, "KG" )
if z>10:
   print("Su pedido supera los 10 KG por un peso de ", z-10, "KG")
```
